```python
#Imports & Load Cleaned Data
import pandas as pd

df = pd.read_csv("../data/sales_cleaned.csv", parse_dates=["Order Date"])
```

```python
# Feature Engineering
df["order_month"] = df["Order Date"].dt.to_period("M").astype(str)
df.head()
```

```python
#Product-wise Sales
product_sales = (
    df.groupby("Product")["Sales"]
    .sum()
    .sort_values(ascending=False)
)
product_sales
```

```python
#Category-wise Sales
category_sales = df.groupby("Category")["Sales"].sum()
category_sales
```

```python
#Region-wise Sales
region_sales = df.groupby("Region")["Sales"].sum()
region_sales
```

```python
#Monthly Sales Trend
monthly_sales = (
    df.groupby("order_month")["Sales"].sum().reset_index())
monthly_sales
```

```python
#Save Aggregated Data
monthly_sales.to_csv("../data/monthly_sales.csv", index=False)
```
